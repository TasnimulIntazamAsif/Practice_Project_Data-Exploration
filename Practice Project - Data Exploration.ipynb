## About Dataset
This classic dataset contains the prices and other attributes of almost 54,000 diamonds. It's a great dataset for beginners learning to work with data analysis and visualization.

### Attributes

* $price =$ price in US dollars (\\$326--\\$18,823)
* $carat =$ weight of the diamond (0.2--5.01)
* $cut =$ quality of the cut (Fair, Good, Very Good, Premium, Ideal)
* $color =$ diamond colour, from J (worst) to D (best)
* $clarity =$ a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
* $x =$ length in mm (0--10.74)
* $y =$ width in mm (0--58.9)
* $z =$ depth in mm (0--31.8)
* $depth =$ total depth percentage = $z / mean(x, y) = 2 * z / (x + y) (43--79)$
* $table =$ width of top of diamond relative to widest point (43--95)

## 2.0 Univariate Analysis

### 2.1 For continuous variables: mean, median, mode, min, max, histograms (distribution), and boxplots

```python
diamonds.describe(include=['category'])
```

```python
mean = diamonds['carat'].mean()
median = diamonds['carat'].median()
mode = diamonds['carat'].mode().values[0]
min_val = diamonds['carat'].min()
max_val = diamonds['carat'].max()

print(f'Mean: {mean:.2f}')
print(f'Median: {median:.2f}')
print(f'Mode: {mode:.2f}')
print(f'Min: {min_val:.2f}')
print(f'Max: {max_val:.2f}')

plt.hist(diamonds['carat'])
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='g', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='b', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')
plt.legend(loc='upper right')
plt.show()

```

We can see that mean is to the right of median and mode. mode or the peek of our distribution is 0.3 which means it is the most occured value or dimonds with a carat of 0.3. we can Also see the the median is less the the mean which means that ouliers exist to the right of the mean and the distribution is right skewed. In order to identify the outliers let's use a box plot.


```python
import seaborn as sns

mean = diamonds['carat'].mean()
median = diamonds['carat'].median()
mode = diamonds['carat'].mode().values[0]

sns.boxplot(x='carat', data=diamonds)
plt.title(f'Boxplot of Carat (Mean: {mean:.2f}, Median: {median:.2f}, Mode: {mode:.2f})')
plt.show()

```

we can see that 50% of the data points lies within caret of 0.5 and 0.90 and both the quartiels are pretty evenly spred out also, the median is 0.70. now we can see that byoond 1.49 which is the upper tail alot of the outliers exists. Also we know that the mode is 0.30 which is less then Q1 thus verifying the skewness of the dataset.

### 2.2 For categorical variables: frequency table, count plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Convert categorical columns to string type
for column in ['cut', 'color', 'clarity']:
    diamonds[column] = diamonds[column].astype(str)

# Function to create countplot and display mean, median, mode
def create_countplot_with_stats(column_name, title):
    colors = sns.color_palette("husl", len(diamonds[column_name].unique()))
    plt.figure(figsize=(8, 6))
    sns.countplot(x=column_name, data=diamonds, palette=colors)
    plt.title(title)
    plt.xlabel(column_name.capitalize())
    plt.ylabel("Count")
    plt.show()


    print(f"Statistics for {column_name.capitalize()}:")
    # Mean, Median not applicable for categorical data
    # Mode is displayed
    mode = diamonds[column_name].mode()[0]
    print(f"  Mode(peak of distribution or highest occurence): {mode}")
    print("-" * 20)  # Separator

# Generate count plots and statistics
create_countplot_with_stats('cut', 'Count of Diamonds by Cut')
create_countplot_with_stats('color', 'Count of Diamonds by Color')
create_countplot_with_stats('clarity', 'Count of Diamonds by Clarity')
```

### 3.2 Categorical & Categorical: stacked column chart

```python

import matplotlib.pyplot as plt

# Create a new dataframe for the stacked column chart
stacked_data = diamonds.groupby(['cut', 'color']).size().unstack()
stacked_data.plot(kind='bar', stacked=True)
plt.title('Stacked Column Chart of Cut vs Color')
plt.show()

stacked_data = diamonds.groupby(['cut', 'clarity']).size().unstack()
stacked_data.plot(kind='bar', stacked=True)
plt.title('Stacked Column Chart of Cut vs Clarity')
plt.show()

stacked_data = diamonds.groupby(['color', 'clarity']).size().unstack()
stacked_data.plot(kind='bar', stacked=True)
plt.title('Stacked Column Chart of Color vs Clarity')
plt.show()

```

 This section explores the relationships between categorical variables in the diamonds dataset using stacked column charts.


### 3.3 Categorical & Continuous: boxplots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create box plots for cut vs carat, depth, table, price, x, y, z
for col in ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']:
    plt.figure(figsize=(10,6))
    sns.boxplot(x='cut', y=col, data=diamonds)
    plt.title(f'Boxplot of {col} vs Cut')
    plt.show()

# Create box plots for color vs carat, depth, table, price, x, y, z
for col in ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']:
    plt.figure(figsize=(10,6))
    sns.boxplot(x='color', y=col, data=diamonds)
    plt.title(f'Boxplot of {col} vs Color')
    plt.show()


```

### 3.4 Correlation Analysis: correlation matrix

```python
correlation_matrix = diamonds.select_dtypes(include=['int64', 'float64']).corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', mask=mask, square=True)
plt.title('Correlation Matrix')
plt.show()



```

The correlation matrix analysis provides insights into the relationships between numerical variables in the dataset. It helps identify strong correlations between variables, which can be useful in feature engineering and model selection.
